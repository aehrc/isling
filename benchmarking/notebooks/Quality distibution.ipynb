Here I look at the distibution of the quality of the reads for 300 whole viral integrations with clean junctions. Reads 150bp long were simulated with x20 coverage and a mean fragment size of 400 with a standard deviation of 15. This is useful for determining the read quality which shoul be filtered

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
header = [0,1,2,5,8,12,16,20,25,30,35,40,50,60,61]
results = [352724, 90241, 87366, 70253, 38968,32475, 23362, 15736, 14626, 12781, 11960,11311,11013, 10845,0] 
plt.plot(header,results)
plt.xlabel("Read MAPQ filtered")
plt.ylabel("Number of reads present")
plt.show()
```
