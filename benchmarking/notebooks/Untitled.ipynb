```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
conf_mat = np.array([[1,2],[4,5]])
axis_labels = ["Viral", "Non-viral"]
```

```python

axis_labels = ["Viral", "Non-viral"]
ax = sns.heatmap(conf_mat, annot= conf_mat, fmt = '.1f',xticklabels = axis_labels,
                yticklabels = axis_labels, annot_kws = {"size":15}, cmap = 'Blues')
for t in ax.texts: t.set_text(t.get_text()+"%")
plt.pyplot.title("Confusion Matrix",fontsize = 20)
plt.pyplot.xlabel('Pipeline Predicted',fontsize = 15)
plt.pyplot.ylabel('Acutal', fontsize = 15)
plt.pyplot.show()
```
